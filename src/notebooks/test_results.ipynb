```python
#%matplotlib inline
import numpy as np
import cPickle as pickle
import matplotlib.pyplot as plt
```

```python
def load_test(filename):
    with open(filename, 'rb') as f:
        model_hist = pickle.load(f)
    epochs = model_hist['epoch_reached']
    train_acc = np.asarray(model_hist['train_accs'])
    test_acc = np.asarray(model_hist['test_accs'])
    disc_enabled = model_hist['disc_enabled']
    learning_rate = model_hist['learning_rate']
    
    return epochs, test_acc
```

```python
test_name = 'non-uniform.model'
epochs_t1, test_acc_t1 = load_test(test_name)

test_name = 'no-quant.model'
epochs_t2, test_acc_t2 = load_test(test_name)
```

```python
plt.figure(1)
plt.plot(range(0, epochs_t1+1), test_acc_t1)
plt.hold(True)
plt.plot(range(0, epochs_t2+1), test_acc_t2)
plt.legend(['Non-Uniform Qnt. Test', 'No Qnt. Test.'], loc=4)
plt.title(test_name)
plt.grid(True)
plt.ylim([.955, .975])
plt.show()

print "Non-Uniform Mean {0:.5f}, Std:{1:.5f}".format(np.mean(test_acc_t1), np.std(test_acc_t1))
print "No Quant Mean {0:.5f}, Std:{1:.5f}".format(np.mean(test_acc_t2), np.std(test_acc_t2))
```
